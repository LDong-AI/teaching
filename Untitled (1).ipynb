计算相关系数：pandas中DataFrame对象corr()方法的用法

```python
import pandas as pd
# 引入seaborn库，为了画相关结果的热力图
import seaborn as sns

data = pd.read_csv("StudentsPerformance.csv")
df = pd.DataFrame(data)   # The pandas object holding the data.
#print(df.shape)

#提取math score；reading score；writing score 
score_data=pd.DataFrame({'math_score':df['math score'],'reading_score':df['reading score'],'writing_score':df['writing score']})
dcorr=score_data.corr()
display(dcorr) # 表格形式显示数据
```

```python
#画热力图（相关系数）
plt.figure(figsize=(4,3),dpi=80)  # dpi调整图像大小
sns.heatmap(data=dcorr,vmax=1,vmin=0.7,cmap="OrRd") #vmax,vmin 是色柱最大值; cmap调整颜色
```

```python
import statsmodels.api as sm
#from statsmodels.sandbox.regression.predstd import wls_prediction_std
                      # 小写的 ols 函数才会自带截距项，OLS 则不会
         # 固定格式：因变量 ~ 自变量(+ 号连接)
#lm = ols('math score ~  reading score + bedrooms + bathrooms', data=df).fit()
#lm.summary()
#先通过 sm.add_constant() 向矩阵 X 添加截距列后，再用 sm.OLS() 建立普通最小二乘模型，
#最后用 model.fit() 就能实现线性回归模型的拟合，并返回拟合与统计分析的结果摘要。
x1=df['writing score']
X = sm.add_constant(x1) 
model=sm.OLS(df['reading score'],X)
results = model.fit()
print(results.summary()) 
```

```python
print("OLS model: Y = b0 + b1 * x") # b0: 回归直线的截距，b1: 回归直线的斜率
print('Parameters: ', results.params) # 输出：拟合模型的系数
yFit = results.fittedvalues # 拟合模型计算出的 y值
fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(x1, df['reading score'], 'o', label="data") # 原始数据
ax.plot(x1, yFit, 'r-', label="OLS") # 拟合数据
```

简单回归:
第一步：分类变量数据初始化。

```python
import numpy as np
Pedu=np.array(df['parental level of education'])
for i in range(len(Pedu)):
    if Pedu[i]=="some high school":
        Pedu[i]=2
    elif Pedu[i]=="high school":
        Pedu[i]=4
    elif Pedu[i]=="associate's degree":
        Pedu[i]=6
    elif Pedu[i]=="some college":
        Pedu[i]=8
    elif Pedu[i]=="bachelor's degree":
        Pedu[i]=10
    else:
        Pedu[i]=12
```

```python
df['parental level of education']=Pedu
df
```

设置虚拟变量

```python
# 名义变量：test preparation course；虚拟变量：nominal_data
nominal_data_tpc=df['test preparation course']
nominal_data_gen=df['gender']
nominal_data_lunch=df['lunch']

# 设置虚拟变量  # pandas会自动帮你命名虚拟变量
dummies1=pd.get_dummies(nominal_data_tpc)  # 自动设置test preparation course的虚拟变量
dummies2=pd.get_dummies(nominal_data_gen)  # 自动设置gender的虚拟变量
dummies3=pd.get_dummies(nominal_data_lunch) 

# 每个名义变量生成的虚拟变量中，需要各丢弃一个，这里以丢弃none为例
dummies1.drop(columns=['none'],inplace=True)
# 将虚拟变量设置结果与原数据集拼接
results = pd.concat(objs=[df,dummies1,dummies2,dummies3],axis='columns') # 按照列来合并

results=results.drop(columns=['test preparation course','gender','female'])
results
```

```python
# 名义变量：test preparation course；虚拟变量：nominal_data
nominal_data=df['race/ethnicity']
# 设置虚拟变量
dummies=pd.get_dummies(nominal_data)  # pandas会自动帮你命名虚拟变量

# 每个名义变量生成的虚拟变量中，需要各丢弃一个，这里以丢弃none为例
dummies.drop(columns=['group E'],inplace=True)
# 将虚拟变量设置结果与原数据集拼接
results = pd.concat(objs=[results,dummies],axis='columns') # 按照列来合并
results=results.drop(columns=['race/ethnicity','lunch'])

results
```

```python
#因为ols函数不允许数据的字段中有空格存在，所以先给各列改名字，去掉列名中的空格
#results.columns = ['ParentalLevelofEducation','MathScore','ReadingScore','WritingScore','Completed','Male','Free/reduced','Standard','groupA','groupB','groupC','groupD']
results.columns = ['PLE','MathScore','ReadingScore','WritingScore','Completed','Male','Free/reduced','Standard','groupA','groupB','groupC','groupD']
results
```

```python
results[['PLE']]= results['PLE'].apply(pd.to_numeric)
print(results.dtypes)
```

```python
#import statsmodels.formula.api as smf
#from statsmodels.formula.api import ols
import statsmodels.formula.api as smf

#X = results[['WritingScore', 'Radio']]
y = results['MathScore']
#est = smf.ols(formula='MathScore ~ WritingScore + ParentalLevelofEducation',data=results).fit() # 未优化（1）

#est = smf.ols(formula='MathScore ~ WritingScore + ParentalLevelofEducation+ groupA+ Male+ Standard',data=results).fit() # 优化加入更多变量
#est = smf.ols(formula='MathScore ~ WritingScore + PLE+ groupA+ Male+ Standard',data=results).fit() # 优化加入更多变量，用于去除多重线性自变量举例子
#est = smf.ols(formula='MathScore ~ WritingScore + PLE+  Male ',data=results).fit() # 
est = smf.ols(formula='MathScore ~ WritingScore + Male ',data=results).fit()#结果用于画3D图,最佳模型

y_pred=est.predict()

print(est.summary()) #回归结果
print(est.params) #系数
```
